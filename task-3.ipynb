```python
import pandas as pd

# Sample data
data = {
    "Date": pd.date_range(start="2025-11-01", periods=10, freq="D"),
    "Close": [100, 102, 101, 103, 105, 107, 106, 108, 110, 112]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create an empty list for SMA values
sma_values = []

# Loop through each row
for i in range(len(df)):
    if i < 2:
        # Not enough data for first 2 rows
        sma_values.append(None)
    else:
        # Take current and previous 2 closing prices
        avg = (df.loc[i, "Close"] + df.loc[i-1, "Close"] + df.loc[i-2, "Close"]) / 3
        sma_values.append(avg)

# Add new column
df["SMA_3"] = sma_values

print(df)
```
